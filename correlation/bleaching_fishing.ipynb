```python
import numpy as np
import pandas as pd
import os,sys
```

```python
path_data = os.path.join(os.getcwd(),os.pardir,os.pardir,'data')
bleaching_data_path = os.path.join(path_data,'bleaching')
fishing_data_path = os.path.join(path_data,'fishing_activity')
```

```python
years = np.arange(2012,2019)
early_years = np.arange(2012,2017)
```

```python
region = 'North America' # 'Asia' or 'North America'
# load mean fishing activity data
fishing_data = np.load(os.path.join(fishing_data_path,'mean_fishing_activity_{}.npy'.format(region))) # （2012, 2018）, 7 years
# load bleaching data
high_count = np.load(os.path.join(bleaching_data_path,'high_count_{}.npy'.format(region))) # （2008,2018）, 11 years
whole_count = np.load(os.path.join(bleaching_data_path,'whole_count_{}.npy'.format(region)))
high_mean = np.load(os.path.join(bleaching_data_path,'high_mean_{}.npy'.format(region)))
whole_mean = np.load(os.path.join(bleaching_data_path,'whole_mean_{}.npy'.format(region)))

# get last 7 years data
high_count = high_count[-7:]
whole_count = whole_count[-7:]
high_mean = high_mean[-7:]
whole_mean = whole_mean[-7:]
```

```python
# Do linear regression
from scipy.stats import linregress
# high count
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,high_count)
print('p_value for high count: {}'.format(p_value))
print('r_value for high count: {}'.format(r_value))
# whole count
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,whole_count)
print('p_value for whole count: {}'.format(p_value))
print('r_value for whole count: {}'.format(r_value))
# high mean
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,high_mean)
print('p_value for high mean: {}'.format(p_value))
print('r_value for high mean: {}'.format(r_value))
# whole mean
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,whole_mean)
print('p_value for whole mean: {}'.format(p_value))
print('r_value for whole mean: {}'.format(r_value))
```

```python
# 4 subplots
import matplotlib.pyplot as plt
fontsize = 15
fontsize_title = 18
fig,ax = plt.subplots(2,2,figsize=(15,15))
ax[0,0].scatter(fishing_data,high_count)
# plot linear regression
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,high_count)
ax[0,0].plot(fishing_data,slope*fishing_data+intercept,'r')
ax[0,0].set_title('Number of serious bleaching positions',fontsize=fontsize_title)
ax[0,0].set_xlabel('fishing activity',fontsize=fontsize)
ax[0,0].set_ylabel('bleaching positions number',fontsize=fontsize)
ax[0,1].scatter(fishing_data,whole_count)
# plot linear regression
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,whole_count)
ax[0,1].plot(fishing_data,slope*fishing_data+intercept,'r')
ax[0,1].set_title('Number of bleaching positions',fontsize=fontsize_title)
ax[0,1].set_xlabel('fishing activity',fontsize=fontsize)
ax[0,1].set_ylabel('bleaching positions number',fontsize=fontsize)
ax[1,0].scatter(fishing_data,high_mean)
# plot linear regression
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,high_mean)
ax[1,0].plot(fishing_data,slope*fishing_data+intercept,'r')
ax[1,0].set_title('Mean Value of serious bleaching percentage',fontsize=fontsize_title)
ax[1,0].set_xlabel('fishing activity',fontsize=fontsize)
ax[1,0].set_ylabel('bleaching percentage',fontsize=fontsize)
ax[1,1].scatter(fishing_data,whole_mean)
# plot linear regression
slope, intercept, r_value, p_value, std_err = linregress(fishing_data,whole_mean)
ax[1,1].plot(fishing_data,slope*fishing_data+intercept,'r')
ax[1,1].set_title('Mean Value of bleaching percentage',fontsize=fontsize_title)
ax[1,1].set_xlabel('fishing activity' ,fontsize=fontsize)
ax[1,1].set_ylabel('bleaching percentage',fontsize=fontsize)
save_fig_folder = os.path.join(os.getcwd(),os.pardir,'figures')
if not os.path.exists(save_fig_folder):
    os.makedirs(save_fig_folder)
plt.savefig(os.path.join(save_fig_folder,'{}_bleaching_fishing.png'.format(region)))
plt.show()

```
