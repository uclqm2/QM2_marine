```python
# import Basemap library and matplotlib
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# Create a Basemap instance centered around Shanghai
map = Basemap(projection='lcc', 
              lat_0=31.2304, lon_0=121.4737,  # Central point of Shanghai
              width=100000, height=120000,  # Adjusted for a closer view
              resolution='h')  # 'h' for high resolution

# Draw coastlines, country boundaries, and map boundary.
map.drawcoastlines()
map.drawcountries()
map.drawmapboundary(fill_color='aqua')

# Fill continents and lakes
map.fillcontinents(color='coral',lake_color='aqua')

# Draw rivers (if available in the Basemap data resolution)
map.drawrivers(color='blue')

# Show the map
plt.title("Shanghai City Map")
plt.show()
```
